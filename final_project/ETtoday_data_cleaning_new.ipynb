```python
import pandas as pd
```

```python
# Read data
df = pd.read_excel('ETtoday_news.xlsx')
df
```

```python
# Filter data with date
from datetime import datetime

df = df[(df["date"] > "2017/12/31)") & (df["date"] < "2021/01/01")].reset_index(drop=True)
df
```

```python
# Filter month
import jieba

df['日期'] = df['date'].apply(lambda x:x[:7])
df
```

```python
# Check Error
for idx, s in enumerate(list(df['content'])):
    if type(s) is int:
        print(idx, s)
```

```python
# Filter category
df = df[df.category.isin(["國際", "地方", "大陸", "影劇", "政治" , "法律", "生活", "社會", "體育"])].reset_index(drop=True)

df
```

```python
# Count each category
df_month = pd.DataFrame(df.groupby(['日期', 'category']).agg('count'))

df_month
```

```python
# Manipulate dataframe with pivot table
df_pivot = df_month.pivot_table('title','日期', 'category')

df_pivot = df_pivot.rename(columns={"category":"日期"})

# drop first column
df_pivot = df_pivot.iloc[0:].reset_index()

df_pivot
```

```python
# ["國際", "地方", "大陸", "影劇", "政治" , "法律", "生活", "社會", "體育"]
# Reform the dataframe
df_new = df_pivot[["日期","國際", "地方", "大陸", "影劇", "政治" , "法律", "生活", "社會", "體育"]].reset_index().set_index('index')
df_new = df_new.to_dict('records')
df_new = pd.DataFrame(df_new)
df_new = df_new.fillna(0).set_index("日期")

df_new["總共篇數"] = df_new.sum(axis=1)

df_new = df_new.rename(
    columns=
    {
    '國際': '國際版篇數',
    '地方': '地方版篇數',
    '大陸': '大陸版篇數',
    '影劇': '影劇版篇數',
    '政治': '政治版篇數',
    '法律': '法律版篇數',
    '生活': '生活版篇數',
    '社會': '社會版篇數',
    '體育': '體育版篇數'
              })
df_new
```

```python
df_wordcount = df

df_wordcount['死'] = df_wordcount['content'].apply(lambda x: True if ('死' in x) or ('撞死' in x) or ('傷' in x) or ('送醫' in x) else False)
df_wordcount['男'] = df_wordcount['content'].apply(lambda x: True if '男' in x else False)
df_wordcount['女'] = df_wordcount['content'].apply(lambda x: True if '女' in x else False)
df_wordcount['肇事'] = df_wordcount['content'].apply(lambda x: True if '肇事' in x else False)
df_wordcount['修法'] = df_wordcount['content'].apply(lambda x: True if '修法' in x else False)
df_wordcount['機車'] = df_wordcount['content'].apply(lambda x: True if ('機車' in x) or ('騎士' in x) or ('騎車' in x) else False)


df_wordcount
```

```python
# 死(撞死、傷、送醫)、男、女、肇事、修法、機車(騎士、騎車)
df_wordcount = df_wordcount.groupby(['日期'])[['死', '男', '女', '肇事', '修法', '機車']].agg('sum')
df_wordcount
```

```python
df_merge = df_new.join(df_wordcount)
df_merge
```

```python
df_output = df_merge.rename(columns={"死":"死(撞死、傷、送醫)", '機車': '機車(騎士、騎車)'})
df_output
```

```python
writer = pd.ExcelWriter('ETtoday_categorized_content.xlsx', engine='xlsxwriter')
df_output.to_excel(writer, sheet_name='Sheet1', index=True)
    
writer.save()
```
